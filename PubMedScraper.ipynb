## Import libraries

```python
from bs4 import BeautifulSoup
import requests
import regex as re
import pandas as pd
import time
import pickle
import numpy as np
```

## Pubmed scraping

### Collecting article pmids

```python
def check_pmids_page(search_url):
    all_pmids_on_page = []
    response = requests.get(search_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    pmids = soup.find_all('span', {'class' : 'docsum-pmid'})
    for p in pmids:
        p = p.get_text()
        all_pmids_on_page.append(p) if p not in all_pmids_on_page else print('pmid already in list')
    return all_pmids_on_page
```

```python
def collect_pmids(search_url_list):
    all_pmids = []
    page_counter = 1
    search_url_index = 0
    while len(all_pmids) < 12500:
        page_pmids = check_pmids_page(search_url_list[search_url_index] + str(page_counter))
        for i in page_pmids:
            if i not in all_pmids and page_counter != 1000: #After visiting 1000 pages the function stalls and does
                all_pmids.append(i) #not add pmids to the list anymore                 
                page_counter += 1
            elif page_counter == 1000:
                search_url_index += 1
                page_counter -= 999
        print(len(all_pmids) / 12500 * 100) #show the progress on collecting 12499 PubMed IDs
        print(page_counter) #Count the page at which it is at
    return(all_pmids)   
```

```python
url = ['https://pubmed.ncbi.nlm.nih.gov/?term=data+science&page=', 
       'https://pubmed.ncbi.nlm.nih.gov/?term=artificial+intelligence&page=',
      'https://pubmed.ncbi.nlm.nih.gov/?term=statistics&page=', 
      'https://pubmed.ncbi.nlm.nih.gov/?term=machine+learning&page=', 
      'https://pubmed.ncbi.nlm.nih.gov/?term=data+statistics&page=', 
      'https://pubmed.ncbi.nlm.nih.gov/?term=probability+theory&page=', 
      'https://pubmed.ncbi.nlm.nih.gov/?term=deep+learning&page=', 
      'https://pubmed.ncbi.nlm.nih.gov/?term=information+science&page=', 
      'https://pubmed.ncbi.nlm.nih.gov/?term=computer+science&page=',
      'https://pubmed.ncbi.nlm.nih.gov/?term=data+mining&page=',
      'https://pubmed.ncbi.nlm.nih.gov/?term=data+engineering&page=',
      'https://pubmed.ncbi.nlm.nih.gov/?term=data+visualisation&page=',
      'https://pubmed.ncbi.nlm.nih.gov/?term=business+intelligence&page=']
pmidsLarge = collect_pmids(url)
```

```python
### Save pmids to file
with open('pmidsLarge', 'wb') as fp:
    pickle.dump(pmidsLarge, fp)
```

```python
### Open file with pmids
with open('pmidsLarge', 'rb') as fp:
    pmidsLarge = pickle.load(fp)
```

### Collecting titles, pmids, urls and abstracts

```python
def collect_info(pmids):
    out = []
    for pmid in pmids:
        url = 'https://pubmed.ncbi.nlm.nih.gov/' + pmid
        
        response = requests.get(url)
        
        soup = BeautifulSoup(response.text, 'html.parser')
        if soup.find("div", {"id": "abstract"}) is not None:
            title = soup.select('h1.heading-title')[0].text.strip()
            abstract = soup.find("div", {"id": "abstract"}).text.replace('\n', '').strip()
            abstract = " ".join(abstract.split()).replace('Abstract', '') 
            data = {'title': title, 'pmid': pmid, 'url':url, 'abstract': abstract}
            time.sleep(3)
            out.append(data)
        print(len(out)/ len(pmids) * 100)
    return(out)
```

```python
df_data = collect_info(pmidsLarge)
```

```python
df = pd.DataFrame(df_data)
```

```python
df.to_csv('title_abstract_data_Large.csv', sep=';', index=False,header=True)
```
